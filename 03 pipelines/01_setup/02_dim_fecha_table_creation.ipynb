```python
%sql
CREATE OR REPLACE TABLE retail.gold.dim_fecha AS
SELECT
  date_format(d, 'yyyyMMdd') AS date_key,
  d AS fecha,
  year(d) AS year,
  date_format(d, 'MMMM') AS mes_full,
  date_format(d, 'MMM') AS mes_short,
  date_format(d, 'MM') AS mes,
  date_format(d, 'dd') AS dia_mes,
  date_format(d, 'E') AS dia_semana_abreviado,
  CASE WHEN date_format(d, 'E') IN ('Sun', 'Sat') THEN 'Y' ELSE 'N' END AS es_fin_de_semana,
  concat('Q', quarter(d)) AS quarter,
  concat(year(d), '-Q', quarter(d)) AS year_quarter
FROM (
  SELECT explode(
    sequence(
      to_date('1998-01-01'),
      to_date('2025-12-31'),
      interval 1 day
    )
  ) AS d
);
```
